```python
# Connexion √† la base de donn√©es et liste des patients
import mysql.connector
from dotenv import load_dotenv
import os

# Charger les variables d'environnement
load_dotenv()

# Configuration de la connexion
DB_CONFIG = {
    'host': os.getenv("DB_HOST"),
    'user': os.getenv("DB_USER"),
    'password': os.getenv("DB_PASS"),
    'database': os.getenv("DB_NAME"),
    'charset': 'utf8mb4'
}

print("="*70)
print("üîå CONNEXION √Ä LA BASE DE DONN√âES")
print("="*70)

try:
    # Connexion
    conn = mysql.connector.connect(**DB_CONFIG)
    cursor = conn.cursor(dictionary=True)
    
    print("‚úÖ Connexion r√©ussie!")
    print(f"üìä Base de donn√©es: {DB_CONFIG['database']}")
    print(f"üñ•Ô∏è  Serveur: {DB_CONFIG['host']}")
    print()
    
    # V√©rifier si la table staging_patients existe
    cursor.execute("SHOW TABLES LIKE 'staging_patients'")
    table_exists = cursor.fetchone()
    
    if table_exists:
        # Compter le nombre de patients
        cursor.execute("SELECT COUNT(*) as total FROM staging_patients")
        count = cursor.fetchone()['total']
        print(f"üìã Nombre total de patients dans staging_patients: {count}")
        print()
        
        if count > 0:
            # Lister les patients (limite √† 10 pour l'affichage)
            cursor.execute("""
                SELECT 
                    id, 
                    first_name, 
                    last_name, 
                    birthdate, 
                    gender,
                    report_filename
                FROM staging_patients 
                ORDER BY id 
                LIMIT 10
            """)
            
            patients = cursor.fetchall()
            
            print("="*70)
            print("üë• LISTE DES PATIENTS (10 premiers)")
            print("="*70)
            
            for i, patient in enumerate(patients, 1):
                print(f"\n{i}. Patient ID: {patient.get('id', 'N/A')}")
                print(f"   Nom: {patient.get('first_name', '')} {patient.get('last_name', '')}")
                print(f"   Date de naissance: {patient.get('birthdate', 'N/A')}")
                print(f"   Sexe: {patient.get('gender', 'N/A')}")
                print(f"   Fichier source: {patient.get('report_filename', 'N/A')}")
            
            if count > 10:
                print(f"\n... et {count - 10} autre(s) patient(s)")
        else:
            print("‚ö†Ô∏è  La table staging_patients est vide")
    else:
        print("‚ö†Ô∏è  La table staging_patients n'existe pas")
        print("   V√©rifiez que le sch√©ma SQL a √©t√© ex√©cut√©")
    
    # Fermer la connexion
    cursor.close()
    conn.close()
    print("\n" + "="*70)
    print("‚úÖ Connexion ferm√©e")
    print("="*70)
    
except mysql.connector.Error as err:
    print(f"‚ùå ERREUR de connexion: {err}")
    print("\nV√©rifiez vos param√®tres dans le fichier .env:")
    print("  - DB_HOST")
    print("  - DB_USER")
    print("  - DB_PASS")
    print("  - DB_NAME")
except Exception as e:
    print(f"‚ùå ERREUR: {e}")

```
